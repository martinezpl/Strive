### To edit an image making a copy of it is always a good idea!

```python
# Load an image
# And make a copy of the original image to avoid manipulating the original image
pic_org = cv2.imread("pub.jpg")
```

[Hint](# "Numpy .copy function") 

## Drawing

### Drawing lines

```python
import matplotlib.pyplot as plt
```

```python
# Draw some lines on top of the image 
pic = pic_org.copy()
pts = np.array([[0, 0], [900, 0], [450, 100], [450, 500]], 
               np.int32) 
  
cv2.polylines(pic, [pts], color=[255,0,0], thickness=10, isClosed=True)
plt.figure(figsize=(10,10))
plt.imshow(pic)
```

[Hint](# "cv2.line function") 

### Circles

```python
# Draw some circles on top of the image
pic = pic_org.copy()
cv2.circle(pic, center=(290, 300), radius=75, color = [255, 0, 0], thickness=5)
cv2.circle(pic, center=(490, 300), radius=75, color = [255, 0, 0], thickness=5)
plt.figure(figsize=(10,10))
plt.imshow(pic)
```

[Hint](# "cv2.circle") 

### Ellipses

```python
# Draw some elipses on top of the image
pic = pic_org.copy()
cv2.ellipse(pic, box=((490,310), (125, 90), 80), color = [255,0,0], thickness=5)
cv2.ellipse(pic, box=((490,310), (90, 300), 180), color = [255,0,0], thickness=5)
cv2.ellipse(pic, box=((490,310), (200, 600), 95), color = [0,255,0], thickness=5)
plt.figure(figsize=(10,10))
plt.imshow(pic)
```

[Hint](# "cv2.ellipse") 

### Poligons

```python
# Lets draw some poligons, first declare the points of the poligon
points = np.array([[300,300], [100, 700],[420,200], [700, 600], [600, 100]])
```

[Hint](# "Example: np.array([[10,5],[120,130],[170,120],[250,210]], np.int32)") 

```python
# Draw the poligon on top of the image
pic = pic_org.copy()
cv2.fillConvexPoly(pic, points, color=[255,0,0])
plt.figure(figsize=(10,10))
plt.imshow(pic)
```
