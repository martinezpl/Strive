```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Template Matching

```python
scene = cv2.imread('img/WaldoScene2.png')
```

```python
main = cv2.imread('img/WaldoScene2.png', 0)
templ = cv2.imread('img/waldo.jpg', 0)

plt.imshow(templ, cmap='gray')
```

```python
match = cv2.matchTemplate(main, templ, cv2.TM_CCORR_NORMED)

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(match)

top_left = max_loc
bottom_right = (top_left[0] + 50, top_left[1] + 50)
cv2.rectangle(scene, top_left, bottom_right, (0, 255, 0), 10)

plt.figure(figsize=(20,20))
plt.imshow(cv2.cvtColor(scene,cv2.COLOR_BGR2RGB))
```
