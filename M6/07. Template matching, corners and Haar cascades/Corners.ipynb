```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
image_g = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris
corners = cv2.cornerHarris(image_g, )

# Display the image with the detected corners 

```

```python
# Convert img to grayscale
image_g = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect the corners using goodFeaturesToTrack
max_corners = 1000
quality = 0.0001
distance = 10
corners = cv2.goodFeaturesToTrack(image_g, max_corners, quality, distance)

# Draw the corners on top of img and show it on screen (make sure the corners are visible!)
for corner in corners:
    x, y = corner[0]
    x, y = int(x), int(y)
    cv2.rectangle(image_g, (x-5, y-5), (x+5, y+5), [0,0,100], 10)
    
plt.imshow(image_g, cmap='gray')
```
