# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Histograms

```python
flowers = cv2.imread('img/blue-flowers.jpg')
plt.imshow(flowers)
```

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
histogram_b = cv2.calcHist([flowers], [0], None, [255], [0, 255])
histogram_g = cv2.calcHist([flowers], [1], None, [255], [0, 255])
histogram_r = cv2.calcHist([flowers], [2], None, [255], [0, 255])
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colors = ['b', 'g', 'r']
for i, c in enumerate(colors):
    histogram = cv2.calcHist([flowers], [i], None, [255], [0, 255])
    plt.plot(histogram, color = c)
```
