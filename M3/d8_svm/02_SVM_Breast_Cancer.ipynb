***Support Vector Machines***
    Using SVM with Python to predict whether a breast cancer tumor is malignant or benign. Exercise uses numpy, pandas, and scikitlearn and utilizes train test split, SVM, SVC,  and GridSearch to identify the best parameters for prediction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
df_feat = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_feat.info()
```

```python
df_target = pd.DataFrame(cancer['target'],columns=['Cancer'])
df_target
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_feat, df_target, test_size = 0.2)
```

```python
#Train the support vector classifier
from sklearn.svm import SVC
svc = SVC()
svc = svc.fit(X_train, np.ravel(y_train))
```

```python
#Predictions and evaluations
from sklearn.metrics import confusion_matrix
y_pred = svc.predict(X_test)
print('score', svc.score(X_test, y_test))
confusion_matrix(y_test, y_pred)
```

***Everything is being classified into a single class! Use Grid Search to search for the best parameters! (Scikitlearn GridSearchCV)***


```python
#Set parameters
from sklearn.model_selection import GridSearchCV

parameters = {'kernel': ('rbf', 'sigmoid', 'poly', 'linear'), 'C': [10, 20]}
clf = GridSearchCV(svc, parameters)
clf.fit(X_train, np.ravel(y_train))
clf_pred = clf.predict(X_test)
print(confusion_matrix(y_test, clf_pred))
df_res = pd.DataFrame(clf.cv_results_)
```
